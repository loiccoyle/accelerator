```python
from accelerator import *
import numpy as np
```

```python
bending_radius = 10000
n_dipoles = 10
dip = Dipole(bending_radius, 2*np.pi / n_dipoles)
drift = Drift(1000)

lattice = Lattice([dip, drift] * n_dipoles)
```

```python
lattice.plot.top_down()
```

```python
lattice.twiss().plot()
```

```python
lattice.dispersion().plot('h')
```

```python
lattice.tune()
```

```python
lattice.tune(dp=1e-6)
```
