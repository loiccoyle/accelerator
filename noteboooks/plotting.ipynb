```python
from accelerator import Lattice
from accelerator import Dipole, Quadrupole, Drift
```

```python
import numpy as np
```

```python
f = 0.8  # focal length (m)
L = 1.  # drift length (m). Note FODO total length: 2L

half_quad_F = Quadrupole(2.*f)
quad_D = Quadrupole(-f)
drift = Drift(1)
```

```python
FODO = Lattice([half_quad_F, drift, quad_D, drift, half_quad_F, Dipole(1, np.pi/2)])
```

```python
fig, ax = FODO.plot.lattice()
```

```python
fig, ax = FODO.plot.top_down()
```
