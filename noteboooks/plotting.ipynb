```python
import numpy as np
import matplotlib.pyplot as plt
```

# lattice plotting

```python
f = 0.8  # focal length (m)
L = 1.  # drift length (m). Note FODO total length: 2L

half_quad_F = QuadrupoleThin(2.*f)
quad_D = QuadrupoleThin(-f)
drift = Drift(1)
```

```python
FODO = Lattice([half_quad_F, drift, quad_D, drift, half_quad_F, Dipole(1, np.pi/2)])
```

```python
FODO.plot.layout()
```

```python
FODO.plot.top_down()
```

# Transport plotting

```python
FODO = Lattice([half_quad_F, drift, quad_D, drift, half_quad_F])
```

```python
twiss = FODO.twiss()
twiss.plot()
```

```python
coords = FODO.transport([0, 1, 0, 0, 0])
coords.plot()
```

```python
ellipse = Beam().ellipse([2, 0, 1/2])
ellipse.plot()
transported_ellipse = FODO.transport(ellipse)
transported_ellipse.plot()
```

```python
distribution = Beam().match([2, 0, 1/2])
distribution.plot()
transported_distribution = FODO.transport(distribution)
transported_distribution.plot()
```
